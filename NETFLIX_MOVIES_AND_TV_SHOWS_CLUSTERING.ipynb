## 10.Cast

```python
# Create a list of all the actors in the data.
def convert_to_list(data):
    actor_list = {}  
    temp_list = []   
    cast_list = list(data['cast'])
    for i in range(len(cast_list)):
        temp_list.extend(str(cast_list[i]).split(','))
    for i in temp_list:
        if i not in actor_list:
            actor_list[i] = temp_list.count(i)
    return actor_list
```

```python
# Create dataframe for list of actors.
actor_list = pd.DataFrame([convert_to_list(Netflix_data)]).T.reset_index()
```

```python
# after above all the changes, those features are in list format, so making list of description feature
Netflix_data['description'] = Netflix_data['description'].apply(lambda x: x.split(' '))
```

```python
# converting text feature to string from list
Netflix_data['description']= Netflix_data['description'].apply(lambda x: " ".join(x))
# making all the words in text feature to lowercase
Netflix_data['description']= Netflix_data['description'].apply(lambda x: x.lower())
```

```python
# create a function for remove punctuation
def remove_punctuation(text):
    '''a function for removing punctuation'''
    import string
    # replacing the punctuations with no space,which in effect deletes the punctuation marks 
    translator = str.maketrans('', '', string.punctuation)
    # return the text stripped of punctuation marks
    return text.translate(translator)
```

```python
# applying above function on text feature
Netflix_data['description']= Netflix_data['description'].apply(remove_punctuation)
```

```python
Netflix_data['description']
```

```python
# using nltk library to download stopwords
sw=stopwords.words('english')

#Defining stopwords 
def stopwords(text):
    '''a function for removing the stopword'''
    text = [word for word in text.split() if word not in sw]
    # joining the list of words with space separator
    return " ".join(text)
```

```python
# applying above function on text feature
Netflix_data['description']=Netflix_data['description'].apply(stopwords)
```

```python
# this is how value in text looks like after removing stopwords
Netflix_data['description']
```

```python
#Applying Tfidf Vectorizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
tfidfmodel = TfidfVectorizer(max_features=5000)
X_tfidf = tfidfmodel.fit_transform(Netflix_data['description'])
X_tfidf.shape
```

```python
# convert X into array form for clustering
X = X_tfidf.toarray()
```

## Clustering algorithms

```python
# Importing libraries for clustering
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples
import scipy.cluster.hierarchy as sch
import warnings
warnings.filterwarnings
```

# 1. K Mean Clustering.

```python
# Creating an instance of the model
kmeans = KMeans()
```

### Elbow Method to find optimum k value

```python
#Initializing the list for the values of WCSS 
wcss_list= [] 
  
#Using for loop for iterations from 1 to 30.  
for i in range(1, 30):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(X)  
    wcss_list.append(kmeans.inertia_)  
plt.plot(range(1, 30), wcss_list)  
plt.title('The Elobw Method Graph')  
plt.xlabel('Number of clusters(k)')  
plt.ylabel('wcss_list')  
plt.show()
```

**From both the above plots, we have experimented and chosen 10 as optimal k**

```python
# Lets select optimal value of K as 10
kmeans= KMeans(n_clusters=10, init= 'k-means++',random_state= 42)
#kmeans.fit(X)
y_predict = kmeans.fit_predict(X)
```

```python
#Predict the clusters and evaluate the silhouette score
score = silhouette_score(X, y_predict)
print("Silhouette score is {}".format(score))
```

```python
#davies_bouldin_score of our clusters 
from sklearn.metrics import davies_bouldin_score
davies_bouldin_score(X, y_predict)
```

### Silhouette Score

```python
#sillhoute score of clusters 
sill_score = [] 
for i in range(2,30):
    model = KMeans(n_clusters=i,init ='k-means++',random_state=51)
    model.fit(X)
    y1 = model.predict(X)
    score = silhouette_score(X,y1)
    sill_score.append(score)
    print('For n_clusters: %d, The silhouette_score is: %0.4f'%(i,score))
```

```python
#Plotting Sillhoute's score
plt.figure(figsize =(12, 6))
plt.plot(sill_score,'bs--')
plt.xticks(list(range(0,30)))
plt.grid(),plt.xlabel('Number of cluster')
plt.show()
```
